# Introduction to Strings in JavaScript

Strings are one of the most commonly used data types in JavaScript. They are immutable, which means their values cannot be changed once they are created.

## Defining Strings
In JavaScript, you can define a string using either single (`'`) or double (`"`) quotes. String indexing starts at `0`.

### Key Examples:

1. **Displaying a String**
   ```javascript
   console.log('This is a simple string');
   ```
2. **String Stored in a Variable**
   ```javascript
   let message = 'This is stored in a variable';
   console.log(message);
   ```
3. **Including Quotes Inside Strings**
   ```javascript
   console.log('The cat said \"Hello!\" as it jumped onto the table');
   ```
4. **Concatenating Two Strings**
   ```javascript
   let firstWord = 'good';
   let secondWord = 'morning';
   console.log(firstWord + ' ' + secondWord);  // good morning
   ```
5. **Concatenating Variables and Text**
   ```javascript
   let animal = 'elephant';
   let habitat = 'the savanna';
   console.log('The ' + animal + ' lives in ' + habitat + '.');
   ```
6. **Template Literals for Easier Concatenation**
   ```javascript
   let location = 'the jungle';
   let animalTwo = 'tiger';
   console.log(`The ${animalTwo} prowls through ${location}.`);
   ```
7. **Using Line Breaks in Template Literals**
   ```javascript
   let poem = `The moonlight shines
on the calm sea.
A quiet night.`;
   console.log(poem);
   ```
