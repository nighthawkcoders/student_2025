## Binary Search in Action

```python
def binary_search_trace(arr, target):
    low = 0
    high = len(arr) - 1
    comparisons = 0
    steps = []

    while low <= high:
        mid = (low + high) // 2
        comparisons += 1
        steps.append(f"Compare {target} with arr[{mid}] = {arr[mid]}")

        if arr[mid] == target:
            steps.append(f"Found {target} at index {mid}")
            return True, comparisons, steps
        elif arr[mid] < target:
            steps.append(f"{target} > {arr[mid]} → Search right (low = {mid + 1})")
            low = mid + 1
        else:
            steps.append(f"{target} < {arr[mid]} → Search left (high = {mid - 1})")
            high = mid - 1

    steps.append(f"{target} not found in list.")
    return False, comparisons, steps


# Sorted list
sorted_list = [3, 6, 9, 12, 15, 18, 21, 24, 27, 30, 33]

# Targets to search
targets = [18, 33, 5]

# Run and display results
for target in targets:
    found, comparisons, steps = binary_search_trace(sorted_list, target)
    print(f"\n🔍 Searching for {target} in {sorted_list}")
    for step in steps:
        print(step)
    print(f"Comparisons made: {comparisons}")
    print(f"Target Found: {'Yes' if found else 'No'}")
```

```python
def binary_search_strings(arr, target):
    low = 0
    high = len(arr) - 1
    comparisons = 0
    steps = []

    while low <= high:
        mid = (low + high) // 2
        comparisons += 1
        current = arr[mid]
        steps.append(f"Compare '{target}' with arr[{mid}] = '{current}'")

        if current == target:
            steps.append(f"Found '{target}' at index {mid}")
            return True, comparisons, steps
        elif current < target:
            steps.append(f"'{target}' > '{current}' → Search right (low = {mid + 1})")
            low = mid + 1
        else:
            steps.append(f"'{target}' < '{current}' → Search left (high = {mid - 1})")
            high = mid - 1

    steps.append(f"'{target}' not found in list.")
    return False, comparisons, steps


# Sorted list of words
word_list = ["apple", "banana", "carrot", "dragonfruit", "fig", "grape", 
             "kiwi", "mango", "orange", "peach", "watermelon"]

# Targets to search
word_targets = ["mango", "carrot", "lemon"]

# Run and display results
for word in word_targets:
    found, comparisons, steps = binary_search_strings(word_list, word)
    print(f"\n🔍 Searching for '{word}' in word list:")
    for step in steps:
        print(step)
    print(f"Comparisons made: {comparisons}")
    print(f"Target Found: {'Yes' if found else 'No'}")
```

## Free Response Answers

- Binary search is more efficient because it cuts the list in half with each step. Instead of checking each item one by one like linear search, it quickly narrows down where the target could be. This makes it much faster, especially for big lists. While linear search might take 1,000 steps in a list of 1,000 items, binary search would only need about 10 steps.
- Binary search won’t work if the list isn’t sorted. It depends on knowing which half of the list to search next based on the middle value. In an unsorted list, the comparisons don’t tell you anything useful, so you might miss the target even if it’s there.
- Yes, you could use binary search in a leaderboard if the scores or names are sorted, because you're often looking for a specific player or rank. It could also work in a streaming service if you're looking for an exact movie title in a sorted list. But most search bars let you type part of a name, use filters, or handle spelling mistakes. These flexible searches need other tools, like tries or indexing, which work better than binary search for those cases.
